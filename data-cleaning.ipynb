```python
import pandas as pd
import ast
```

```python
credits = pd.read_csv('credit_data.csv')
movies = pd.read_csv('movie_data.csv')
```

```python
credits.info()
```

```python
movies.info()
```

```python
movies = movies[['id','genres','keywords','title','tagline','overview','production_companies']]
credits = credits[['movie_id','cast','crew']]
```

```python
movies.info()
credits.info()
```

```python
# join both dataframes using merge
table = pd.merge(movies, credits, left_on = 'id', right_on = 'movie_id', how = 'inner')
```

```python
table.drop(columns = ['movie_id'], inplace = True)
```

```python
table.info()
```

```python
table
```

```python
data = table.loc[2,'crew']
data
```

```python
def fetch_data(data):
    a = ast.literal_eval(data)
    data_store = []
    for item in a:
        data_store.append(item['name'])
    comma_data = ','.join(data_store)
    return comma_data
```

```python
table['genres'] = table['genres'].apply(fetch_data)
table['keywords'] = table['keywords'].apply(fetch_data)
table['production_companies'] = table['production_companies'].apply(fetch_data)
table['cast'] = table['cast'].apply(fetch_data)
table['crew'] = table['crew'].apply(fetch_data)
```

```python
table['genres'] = table['genres'].str.replace(' ','_')
table['keywords'] = table['keywords'].str.replace(' ','_')
table['production_companies'] = table['production_companies'].str.replace(' ','_')
table['cast'] = table['cast'].str.replace(' ','_')
table['crew'] = table['crew'].str.replace(' ','_')
```

```python
table
```

```python
table.fillna('',inplace = True)
```

```python
table['tags'] = table['genres']+','+table['keywords']+','+table['tagline']+','+table['overview']+','+table['production_companies']+','+table['cast']+','+table['crew']
```

```python
table = table[['id','title','tags']]
```

```python
table['title'] = table['title'].str.lower()
table['tags'] = table['tags'].str.lower()
```

```python
table
```

```python
table['tags'] = table['tags'].str.replace('.','')
```

```python
table.loc[0,'tags']
```

```python
table = table.drop(table.sample(n=3000, random_state=42).index)
```

```python
table.to_csv('cleaned_data.csv', index = False)
```
