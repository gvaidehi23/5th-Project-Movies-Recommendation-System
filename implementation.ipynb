```python
import pandas as pd
import numpy as np
import pickle
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.info()
```

```python
obj = TfidfVectorizer(max_features = 10000, stop_words = 'english')
vector = obj.fit_transform(df['tags']).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def index_from_name(name):
    movie_name = name.strip().lower().replace(' ','').replace('_','')
    match = df[df['title'].str.strip().str.lower().str.replace(' ','').str.replace('_','') == movie_name ]
    if match.empty:
        return -1   
    return match.index[0]
```

```python
def name_from_index(i):
    if  (i < len(df)) and (i > 0):
        return df.loc[i, 'title']
    return ""
```

```python
op1 = index_from_name('Shanghai Calling')
op2 = name_from_index(4801)
```

```python
op2
```

```python
enter_name = input('Enter movie name:')
fetch_index = index_from_name(enter_name)

if fetch_index != -1:
    similarity_index  = list((enumerate(similarity[fetch_index])))
    similarity_index = sorted(similarity_index, key = lambda x:x[1], reverse = True)
    print('Because u watched ',enter_name)
    print('You must watch following movies:')
    for i in range (1,6):
        final_index = similarity_index[i][0]
        print(i,'. ',name_from_index(final_index))
else:
    print('Movie not found')
```

```python
pickle.dump(similarity, open('model.pkl','wb'))
```
